# Handling the inappropriate Data types for the above Dataset- Data Scrubbing

```python
#convert 'dteday' to datetime type

df['dteday'] = pd.to_datetime(df['dteday'], format="%Y-%m-%d")
```

```python
#Adding a new coloumn for date and time combiningly 
df['datetime'] = df['dteday'] + pd.to_timedelta(df['hr'], unit='h')
```

```python
#Ensure categorical values are treated as such

categorical_vars = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
df[categorical_vars] = df[categorical_vars].astype('category')
```

```python
#drop instant
df.drop('instant', axis=1, inplace=True)
```

```python
#Create 'rush_hour' feature based on typical rush_hour times in the morning and evening 
df['rush_hour'] = ((df['hr']>=8)&(df['hr']<9) | (df['hr']>=17)&(df['hr']<=18)).astype(int)

#display the first few rows to verify the new feature
print(df[['hr','rush_hour']].head(10))
```

```python
#Box plot of count of rentals durting rush hours vs non-rush hours
plt.figure(figsize=(8,6))
sns.boxplot(x='rush_hour',y='cnt',data=df)
plt.title('Box plot of Rental Counts by Rush Hour vs Non-Rush Hour')
plt.xlabel('Rush Hour(1:Yes, 0:No)')
plt.ylabel('Count of Rentals')
plt.xticks([0,1],['Non-Rush Hour','Rush Hour'])            
plt.show()
```

```python
#Create interaction variable for 'rush hour on working '
df['rush_hour_workingday'] = df['rush_hour'] & df['workingday'].astype('int64')

#Verification of the new added feature
print(df[['hr','rush_hour','workingday','rush_hour_workingday']].head(10))
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='rush_hour_workingday',y='cnt',data=df)
plt.title('Box plot of Rental Counts by Rush Hour(workday) vs Non-Rush Hour')
plt.xlabel('Rush Hour(1:Yes, 0:No)')
plt.ylabel('Count of Rentals')
plt.xticks([0,1],['Non-Rush Hour','Rush Hour'])            
plt.show()
```

**as per plot we notice the later one is a stronger feature with negligible overlapping**

```python
from sklearn.preprocessing import StandardScaler

#Select numerical columns(excluding categorical and datetime columns)
numerical_cols = ['temp','atemp','hum','windspeed']

#Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[numerical_cols])

#Cross-check the standardize data
print(df_scaled[:5])
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(df_scaled)

#Check the explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(explained_variance)

#Cumulative explained variance to decide the number of components
cumulative_explained_variance = np.cumsum(explained_variance)
print(cumulative_explained_variance)
```

```python
# Plot the explained variance
plt.figure(figsize=(8,4))
plt.plot(np.arange(1,5), cumulative_explained_variance)
plt.title('Explained Variance by different Principal components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)           
plt.show()
```

```python
#Assuming we decide to keep the first three components
pca = PCA(n_components=3)
df_pca = pca.fit_transform(df_scaled)

#Checking the shape of transformed data
print(df_pca.shape)
```

```python
#Loading of first 3 components

loadings = pca.components_
print(loadings)
```

```python
from sklearn.preprocessing import StandardScaler

#Assuming the numerical columns as before
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[numerical_cols])

#Optionally select a random subset of data for t-SNE if the Dataset is Large
n_samples = 1000
random_indices = np.random.choice(df_scaled.shape[0], n_samples, replace=False)
data_subset = df_scaled[random_indices]
data_subset.shape
```

```python
from sklearn.manifold import TSNE

#Initialize the t-SNE with two compnents (2D visualization) and a perplexity between 30 and 50
tsne = TSNE(n_components=2, perplexity=30,n_iter=300)
tsne_results = tsne.fit_transform(data_subset)
print(tsne_results.shape)
```

```python
#Color by 'cnt' for visualization of count distribution
plt.figure(figsize=(10,6))
scatter=plt.scatter(tsne_results[:,0], tsne_results[:,1], c=data_subset[:,-1], cmap='viridis', alpha=0.5)
plt.title('t-SNE Plot of Bike rentals')
plt.xlabel('TSNE Component 1')
plt.ylabel('TSNE Component 2')
plt.colorbar(scatter, label='Count of Rentals (cnt)')
plt.show()
```
