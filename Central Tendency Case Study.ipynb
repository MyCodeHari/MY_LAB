**TASK-2**

**Distribution of various features of the data**


*   calculate appropriate measures of central tendency of the features

```python
#Create histogram of the "fixed Acidity" feature
plt.figure(figsize=(11,6))
sns.histplot(data=data, x="fixed acidity", color= "orange",
             edgecolor="linen", alpha=0.5, bins=5)
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.title('Histogram of Fixed Acidity')
plt.show()
```

```python
#Mean of fixed acidity
round(data['fixed acidity'].mean(),2)
```

```python
#Calculate the median of fixed acidity
round(data['fixed acidity'].median(),2)
```

```python
#Create a histogram of "fixed acidity" feature and also show the mean and meadian using seaborn
plt.figure(figsize=(11,6))
sns.histplot(data=data, x="fixed acidity", color= "orange",
             edgecolor="linen", alpha=0.5, bins=5)
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.title('Histogram of Fixed Acidity')
plt.vlines(data['fixed acidity'].mean(),ymin=0,ymax=4000,color='blue', label='Mean')
plt.vlines(data['fixed acidity'].median(),ymin=0,ymax=4000,color='red',label='Median')
plt.legend()
plt.show()
```

**OBSERVATION**:Mean and median are very close to each other so we can use either of the parameters as the value for central tendency

```python
#Create a histogram of "volatile acidity" feature
plt.figure(figsize=(11,6))
sns.histplot(data=data, x="volatile acidity", color= "green",
             edgecolor="linen", alpha=0.5, bins=5)
plt.xlabel('volatile Acidity')
plt.ylabel('Count')
plt.title('Histogram of volatile Acidity')
plt.show()
```

**OBSERVATION**:We have a skewness towards the right of the chart an dis not well distributed, so we can use distplot function to check the distribution

```python
#Create a distplot of "volatile acidity" feature
plt.figure(figsize=(11,6))
sns.distplot(data['volatile acidity'], color= "blue")
plt.xlabel('volatile Acidity')
plt.ylabel('Density')
plt.title('Distplot of volatile Acidity')
plt.show()
```

**OBSERVATION**:It shows a normal distribution also known as "Bell Curve" due to its shape ,described by the mean and the standard deviation curve

*   The mean, median are same/equal with only one mode
*   Symmetric-getting decreased with same amount of both left and right of the centre



```python
#Calculate skewness of "volatile acidity"
round(data['volatile acidity'].skew(),2)
```

**OBSERVATION**:skewness is greater than 1, i.e -postively skewed

```python
#calculate the mean of "volatile acidity"
round(data['volatile acidity'].mean(),2)
```

```python
#calulate the median of "volatile acidity"
round(data['volatile acidity'].median(),2)
```

```python
#Create a histogram of "volatile acidity" feature and also show the mean and meadian.
plt.figure(figsize=(11,6))
sns.histplot(data=data, x="volatile acidity", color= "green",
             edgecolor="linen", alpha=0.5, bins=5)
plt.xlabel('volatile Acidity')
plt.ylabel('Density')
plt.title('Histogram of volatile Acidity')
plt.vlines(data['volatile acidity'].mean(),ymin=0,ymax=4000,color='blue', label='Mean')
plt.vlines(data['volatile acidity'].median(),ymin=0,ymax=4000,color='red',label='Median')
plt.legend()
plt.show()
```

**OBSERVATION**:


*   the mean and median are very close to each with a very small difference
*   so, we can choose mean as the measure of central tendency here



```python
#Create a histogram for "citric acid" feature
plt.figure(figsize=(11,6))
sns.histplot(data=data, x="citric acid", color= "red",
             edgecolor="linen", alpha=0.5, bins=5)
plt.xlabel('citric acidity')
plt.ylabel('Count')
plt.title('Histogram of Citric Acidity')
plt.show()
```

**OBSERVATION**:not well distributed, skewed much towards right

```python
#Calulate the mean of the "Citric acid "feature
round(data['citric acid'].mean(),2)
```

```python
#Calculate the median of the "citic acid" feature
round(data['citric acid'].median(),2)
```

```python
#Create a histogram for "citric acid" feature and also show the mean and median on the chart
plt.figure(figsize=(11,6))
sns.histplot(data=data, x="citric acid", color= "red",
             edgecolor="linen", alpha=0.5, bins=5)
plt.xlabel('citric acidity')
plt.ylabel('Count')
plt.title('Histogram of Citric Acidity')
plt.vlines(data['citric acid'].mean(),ymin=0,ymax=4000,color='blue', label='Mean')
plt.vlines(data['citric acid'].median(),ymin=0,ymax=4000,color='red',label='Median')
plt.legend()
plt.show()
```

**OBSERVATION**:
*   Here also we have our mean and median very close to themselves, so we can choose the mean as the measure of central tendency

```python
#Create a distplot of "Citric acidity" feature
plt.figure(figsize=(11,6))
sns.distplot(data['citric acid'], color= "blue")
plt.xlabel('Citric Acid')
plt.ylabel('Density')
plt.title('Distplot of citric Acidity')
plt.show()
```

**OBSERVATION**:normally distributed plot

*   Nearly same/equal mean and median with only one mode
*   Symmetric, i.e: evenly distribution towards left and right with respect to the centre

**TASK-3**: Create a new pandas series that contains the details of acid types for a quality  

*   Representative value for each of the features



```python
#create a pandas series called "rep_acid" that contains the details of the representative quality for the diffrenet types of acids
rep_acid=pd.Series(index=['fixed acidity','volatile acidity','citric acid','quality'],
                   data=[data['fixed acidity'].mean(),data['volatile acidity'].mean(),data['citric acid'].mean(),
                         data['quality'].value_counts().index[0]])
print(rep_acid)
```
